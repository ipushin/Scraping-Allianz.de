```python
def get_price(raw_price, clean=True):
    if clean:
        raw_price = re.findall(r'(?<=\().+(?=\))', raw_price)[0]
    price = re.sub(r'\,', '.', raw_price)
    price = re.sub(r'[^\.\d]', '', price)
    price = float(price)
    return price
```

```python
# War oder ist Ihr Tier wegen Krankheiten, Fehlentwicklungen oder Unfällen beim Tierarzt in Behandlung?
def treatment_select(treatment):
    
    path_surgery_amount = '//*[@formcontrolname="surgeryAmount"]//input'
    path_diagnosis = '//*[@formcontrolname="preExistingDiagnosis"]//nx-radio[{}]'
    path_treatment = '//*[@formcontrolname="treatments"]//nx-checkbox[{}]'
    
    if treatment == 'Nein':
        # Choosing "Nein"
        browser.find_element_by_xpath(path_diagnosis.format(2)).click()
    
    if treatment == 'Ja & Behandlung ohne Operation':
        # Choosing "Nein" to reset
        browser.find_element_by_xpath(path_diagnosis.format(2)).click()
        # Choosing "Ja"
        browser.find_element_by_xpath(path_diagnosis.format(1)).click()
        # Choosing "Behandlung ohne Operation (Heilbehandlung)"
        browser.find_element_by_xpath(path_treatment.format(1)).click()
        
        
    if treatment == 'Ja & Operationen 1':
        # Choosing "Nein" to reset
        browser.find_element_by_xpath(path_diagnosis.format(2)).click()
        # Choosing "Ja"
        browser.find_element_by_xpath(path_diagnosis.format(1)).click()
        # Choosing "Operationen"
        browser.find_element_by_xpath(path_treatment.format(2)).click()
        
        # Chosing "Anzahl der Operationen"
        browser.find_element_by_xpath(path_surgery_amount).clear()
        browser.find_element_by_xpath(path_surgery_amount).send_keys(1)

    
    if treatment == 'Ja & Behandlung ohne Operation & Operationen 1':
        # Choosing "Nein" to reset
        browser.find_element_by_xpath(path_diagnosis.format(2)).click()
        # Choosing "Ja"
        browser.find_element_by_xpath(path_diagnosis.format(1)).click()
        # Choosing "Behandlung ohne Operation (Heilbehandlung)"
        browser.find_element_by_xpath(path_treatment.format(1)).click()
        # Choosing "Operationen"
        browser.find_element_by_xpath(path_treatment.format(2)).click()
        # Chosing "Anzahl der Operationen"
        browser.find_element_by_xpath(path_surgery_amount).clear()
        browser.find_element_by_xpath(path_surgery_amount).send_keys(1)
        
    
    if treatment == 'Ja & Operationen 2':
        # Choosing "Nein" to reset
        browser.find_element_by_xpath(path_diagnosis.format(2)).click()
        # Choosing "Ja"
        browser.find_element_by_xpath(path_diagnosis.format(1)).click()
        
        # Choosing "Operationen"
        browser.find_element_by_xpath(path_treatment.format(2)).click()
        
        # Chosing "Anzahl der Operationen"
        browser.find_element_by_xpath(path_surgery_amount).clear()
        browser.find_element_by_xpath(path_surgery_amount).send_keys(2)
        
        
    if treatment == 'Ja & Behandlung ohne Operation & Operationen 2':
        # Choosing "Nein" to reset
        browser.find_element_by_xpath(path_diagnosis.format(2)).click()
        # Choosing "Ja"
        browser.find_element_by_xpath(path_diagnosis.format(1)).click()
        
        # Choosing "Operationen"
        browser.find_element_by_xpath(path_treatment.format(2)).click()
        
        # Choosing "Behandlung ohne Operation (Heilbehandlung)"
        browser.find_element_by_xpath(path_treatment.format(1)).click()
        
        # Chosing "Anzahl der Operationen"
        browser.find_element_by_xpath(path_surgery_amount).clear()
        browser.find_element_by_xpath(path_surgery_amount).send_keys(2)
        
       
```

```python
def race_select(race_price_group):
    
    path_race = '//*[@formcontrolname="animalRace"]'
    race_given = browser.find_element_by_xpath(path_race).text
    
    if race_price_group == 1:
        element = WebDriverWait(browser, 10).until(EC.visibility_of_element_located((By.XPATH, path_race)))
        element.click()
        sleep(1)
        
        for _ in range(1):
            sleep(0.1)
            actions.send_keys(Keys.PAGE_DOWN).perform()
        
        
        for n in [10,9]:
            try:
                path = '/html/body/div[5]/div[{}]/div/div/div[2]/nx-dropdown-item[2]'
                element = WebDriverWait(browser, 10).until(EC.visibility_of_element_located((By.XPATH, path.format(n))))
                element.click()
                sleep(0.2)
            except:
                pass

    elif race_price_group == 0 and race_given != 'Mischling / Hauskatze / Europäisch Kurzhaar':
        element = WebDriverWait(browser, 10).until(EC.visibility_of_element_located((By.XPATH, path_race)))
        element.click()
        sleep(1)

        for n in [10,9]:
            try:
                path = '/html/body/div[5]/div[{}]/div/div/div[2]/nx-dropdown-item[1]'
                element = WebDriverWait(browser, 10).until(EC.visibility_of_element_located((By.XPATH, path.format(n))))
                element.click()
            except:
                pass
    
    else:
        pass
```

```python
def birthdate_select(date):
    
    path = '//*[@formcontrolname="animalBirthDate"]'
    browser.find_element_by_xpath(path).clear()

    try:
        browser.find_element_by_xpath(path).send_keys('00.00.00')
        sleep(0.4)
        browser.find_element_by_xpath(path).clear()
        sleep(1)
        browser.find_element_by_xpath(path).send_keys(Keys.ESCAPE)
    except:
        pass

    browser.find_element_by_xpath(path).send_keys(date)
    sleep(0.2)
    browser.find_element_by_xpath(path).send_keys(Keys.ESCAPE)
```

```python
def percent_select(percent):
    # Percentage on the price page
    path = '//*[@formcontrolname="retention"]'
    browser.find_element_by_xpath(path).click()
    
    if percent == '0%':
        path = '/html/body/div[5]/div[4]/div/div/div/nx-dropdown-item[1]/div/div'
        browser.find_element_by_xpath(path).click()
    
    elif percent == '20%':
        path = '/html/body/div[5]/div[4]/div/div/div/nx-dropdown-item[3]/div/div'
        browser.find_element_by_xpath(path).click()
```

```python
def gender_select(gender, waiting_time=10):
    
    # Selecting Gender
    path = '//*[@formcontrolname="animalGender"]'
    element = WebDriverWait(browser, waiting_time).until(EC.element_to_be_clickable((By.XPATH, path)))
    element.click()
    
    sleep(0.6)
    if gender == 'Männlich':
        path = '/html/body/div[5]/div[{}]/div/div/div/nx-dropdown-item[1]/div/div'
        for n in range(9,11):
            try:
                browser.find_element_by_xpath(path.format(n)).click()
            except:
                pass
    
    elif gender == 'Weiblich':
        path = '/html/body/div[5]/div[{}]/div/div/div/nx-dropdown-item[2]/div/div'
        for n in range(9,11):
            try:
                browser.find_element_by_xpath(path.format(n)).click()
            except:
                pass
```

```python
def sterilized_select(sterilized):
    if sterilized == 'Ja':
        # Wurde Ihr Tier kastriert oder sterilisiert? # Ja
        path = '//*[@formcontrolname="sterilized"]//nx-radio[1]'
        browser.find_element_by_xpath(path).click()
    
    elif sterilized == 'Nein':
        # Wurde Ihr Tier kastriert oder sterilisiert? # Nein
        path = '//*[@formcontrolname="sterilized"]//nx-radio[2]' 
        browser.find_element_by_xpath(path).click()
```

```python
def housing_select(housing):
    
    if housing == 'Wohnung':
        # Wie wird Ihre Katze gehalten?
        # Choosing "ausschließlich in der Wohnung"
        path = '//*[@formcontrolname="catHousingType"]//nx-radio[1]'
        browser.find_element_by_xpath(path).click()
    
    elif housing == 'Freigänger':
        # Wie wird Ihre Katze gehalten?
        # Choosing "Freigänger"
        path = '//*[@formcontrolname="catHousingType"]//nx-radio[2]'
        browser.find_element_by_xpath(path).click()
```

```python
def zip_select(zip_group, waiting_time=40):
    zip_code = zip_prices[(zip_prices['ZIP'].isin(unique_zip_groups)) & 
                          (zip_prices['Group Num'] == zip_group)]['ZIP'].values[0]
    
    # Input ZIP
    path = '//*[@formcontrolname="zip"]'
    browser.find_element_by_xpath(path).clear()
    sleep(0.3)
    browser.find_element_by_xpath(path).send_keys(str(zip_code))
    browser.find_element_by_xpath(path).send_keys(Keys.ENTER)


    sleep(0.3)
    # Street
    path = '//*[@formcontrolname="street"]'
    element = WebDriverWait(browser, waiting_time).until(EC.element_to_be_clickable((By.XPATH, path)))
    element.click()
    browser.find_element_by_xpath(path).send_keys(Keys.ARROW_DOWN)
    browser.find_element_by_xpath(path).send_keys(Keys.ENTER)


    sleep(0.3)
    # House number
    path = '//*[@formcontrolname="streetNumber"]'
    browser.find_element_by_xpath(path).send_keys('1')
```

```python
def start_light_browsing():
    
    ua = UserAgent()
    user_agent = ua.random
    proxy = next(proxy_pool)
    
    webdriver.DesiredCapabilities.CHROME['proxy'] = {
        "httpProxy": proxy,
        "ftpProxy": proxy,
        "sslProxy": proxy,
        "proxyType": "MANUAL"
    }
    chrome_options = webdriver.ChromeOptions()
    chrome_options.headless=False
    chrome_options.add_argument('--user-agent=%s' % user_agent)
    
    chrome_options.add_argument("--disable-notifications")
    chrome_options.add_experimental_option("prefs", {"profile.default_content_settings.cookies": 2})
    path = '/Users/macbook/Downloads/Chrome Driver'
    browser =  webdriver.Chrome(os.path.join(path, 'chromedriver'),chrome_options=chrome_options)
    browser.maximize_window()
    
    
    url = 'https://www.allianz.de/gesundheit/tierkrankenversicherung/rechner/#/'
    browser.get(url)
    

    # Cookies
    path = '/html/body/div[1]/div[2]/div[4]/div[2]/div/button'
    element = WebDriverWait(browser, 10).until(EC.element_to_be_clickable((By.XPATH, path)))
    element.click()
    
    # Input ZIP
    path = '//*[@formcontrolname="zip"]'
    element = WebDriverWait(browser, 10).until(EC.element_to_be_clickable((By.XPATH, path)))
    element.clear()

    sleep(0.7)
    # Input birthdate
    path = '//*[@formcontrolname="birthdate"]'
    browser.find_element_by_xpath(path).send_keys('01.12.1990')
    
    # Choose pet
    path = '//*[@formcontrolname="animalCategory"]'
    browser.find_element_by_xpath(path).click()
    sleep(0.4)
    browser.find_element_by_xpath(path).send_keys(Keys.ARROW_DOWN)
    sleep(0.2)
    browser.find_element_by_xpath(path).send_keys(Keys.ENTER)
    
    
    return browser, proxy
```

```python
combinations = pd.read_csv('ALIANZ-COMBINATIONS.csv', index_col=[0])
combinations = combinations.sort_values(['ZIP Price Group','Birth Date','Race Price Group',
                                         'Gender', 'Deduct Percent','Treatment']).reset_index(drop=True)
combined_prices = pd.DataFrame()
```

```python
%%time
sleep_time = 0.3
run = 0
batch = 30
browser, proxy = start_light_browsing()
actions = ActionChains(browser)
print(datetime.now())
file_save = 'ALIANZ_COMBINED.csv'

for idx in combinations.index:
    parameter = combinations['Parameters'].loc[idx]
    
    if idx%10==0 and idx>0:
        combined_prices.to_csv(file_save)
        print(datetime.now())
    
    """
    Selecting parameters
    """
    try:       
        for _ in range(2):
            sleep(0.1)
            actions.send_keys(Keys.PAGE_UP).perform()
        
        # ZIP Price Group
        zip_group = combinations['ZIP Price Group'].loc[idx]

        if run > 0 and zip_group == combinations['ZIP Price Group'].loc[idx-1]:
            pass

        else:
            zip_select(zip_group, waiting_time=7)

        # Gender
        gender = combinations['Gender'].loc[idx]

        if run>0 and gender == combinations['Gender'].loc[idx-1]:
            pass
        else:
            sleep(0.2)
            gender_select(gender)

        # Race
        race_price_group = combinations['Race Price Group'].loc[idx]

        if run>0 and race_price_group == combinations['Race Price Group'].loc[idx-1]:
            pass
        else:
            sleep(0.2)
            race_select(race_price_group)

        # Birth Date
        date = combinations['Birth Date'].loc[idx]

        if run>0 and date == combinations['Birth Date'].loc[idx-1]:
            pass
        else:
            for _ in range(1):
                sleep(0.1)
                actions.send_keys(Keys.PAGE_UP).perform()
            birthdate_select(date)


        # Sterilized
        sterilized = combinations['Sterilized'].loc[idx]

        if run>0 and sterilized == combinations['Sterilized'].loc[idx-1]:
            pass
        else:
            sleep(0.3)
            sterilized_select(sterilized)


        # Housing
        housing = combinations['Housing'].loc[idx]
        if run>0 and housing == combinations['Housing'].loc[idx-1]:
            pass
        else:
            sleep(0.3)
            housing_select(housing)


        # Treatment
        treatment = combinations['Treatment'].loc[idx]

        if run>0 and treatment == combinations['Treatment'].loc[idx-1]:
            pass

        else:
            if treatment !='Nein':
                sleep(0.3)
                treatment_select(treatment)

                sleep(0.3)
                 # Excluding diagnosis
                path_diseases = '//*[@formcontrolname="excludedExistingDiagnosis"]//nx-radio[2]'
                browser.find_element_by_xpath(path_diseases).click()

            else:
                sleep(0.3)
                treatment_select(treatment)
   
        '''
        Getting prices
        '''
   
        callout = 'ZIP: {} | Gender: {} | Race: {} | Date: {} | Sterilized: {} | Housing: {} | Treatment: {}'
        print('\n#{} | Run: {}'.format(idx, run), callout.format(zip_group, gender, race_price_group, date, sterilized, housing, treatment))

        # Go to prices
        sleep(0.3)
        path = '/html/body/div[3]/div/div[4]/div/div/div/div/div/app-root/div/page-data/div/div[2]/form/div/div/div[{}]/button'
        for n in [8,7,6,5]:
            try:
                browser.find_element_by_xpath(path.format(n)).click()
            except:
                pass

        # Deduct Percent
        percent = combinations['Deduct Percent'].loc[idx]
        if run>0 and percent == combinations['Deduct Percent'].loc[idx-1]:
            pass

        else:
            sleep(0.5)
            percent_select(percent)
            sleep(2)

        # Waiting until the price is presented
        sleep(0.3)
        path = '//*[@id="product_TM001"]'
        element = WebDriverWait(browser, 7).until(EC.visibility_of_element_located((By.XPATH, path)))
        
  
        # Get prices
        page_df = populate_page()

        if idx>0 and page_df['Basic'].loc[idx] == combined_prices['Basic'].iloc[-1]:
            print("PRICE ERROR")
            page_df['Correct Price'] = 'No'
            combined_prices = pd.concat([combined_prices, page_df])
            run = batch
           
        else:
            print('GOT DATA')
            page_df['Correct Price'] = 'Yes'
            combined_prices = pd.concat([combined_prices, page_df])
            run+=1
            
        browser.back()
        
    
    except:
        run = batch
        
        print('\n#{}| PAGE ERROR | Run: {} '.format(idx, run), callout.format(zip_group, gender, race_price_group, date, sterilized, housing, treatment))
        try:
            browser.back()
        except:
            pass
    
    if run%batch == 0:

        try:
            browser.quit()
            browser, proxy = start_light_browsing()
            actions = ActionChains(browser)
            run = 0
            
            print('\nNEW SESSION', proxy, run)
            
        except:
            print('\nBROWSER ERROR', proxy, run)
            run = batch - 1
            pass
```
